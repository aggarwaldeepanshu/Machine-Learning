### Count of words in Spam and Non Spam emails

```python
spam_dict={}
not_spam_dict={}
total_words=[]
spam_count=0
total=0
for i in range(0,len(x_words)):
    if(y[i] == "spam"):
        spam_count+=1
    total+=1
    for j in range(0,len(x_words[i])):
        total_words.append(x_words[i][j])
        if(y[i] == "spam"):
            if(x_words[i][j] in spam_dict):
                spam_dict[x_words[i][j]]+=1
            else:
                spam_dict[x_words[i][j]]=1
        else:
            if(x_words[i][j] in not_spam_dict):
                not_spam_dict[x_words[i][j]]+=1
            else:

                not_spam_dict[x_words[i][j]]=1
p_spam=(spam_count)/total
p_notspam=(total-spam_count)/total
unique_words=list(set(total_words))
```

```python
spam_words_size=len(spam_dict)
not_spam_words_size=len(not_spam_dict)
not_spam_words_size
```

### find probability of each word in spam and non spam

```python
for i in range(len(unique_words)):
    if(unique_words[i] in spam_dict):
        spam_dict[unique_words[i]]=spam_dict[unique_words[i]]/spam_words_size
        
    if(unique_words[i] in not_spam_dict):
        not_spam_dict[unique_words[i]]=not_spam_dict[unique_words[i]]/not_spam_words_size

spam_dict['UNK']=0.005
not_spam_dict['UNK']=0.005
    
```

```python
x1_words=list(X_test.words)
y1=list(y_test)
```

### convert ham and spam to 0 and 1 in y_train and y_test

```python
for i in range(len(y)):
    if y[i]=='ham':
        y[i]=0
    else:
        y[i]=1
```

```python
for i in range(len(y1)):
    if y1[i]=='ham':
        y1[i]=0
    else:
        y1[i]=1
```

### find result for each email

```python
result1=1
result2=1
ans=[]
for i in range(len(x1_words)):
    for j in range(len(x1_words[i])):
        if x1_words[i][j] in spam_dict:
            result1*=spam_dict[x1_words[i][j]]
        else:
            result1*=spam_dict['UNK']
    for j in range(len(x1_words[i])):
        if x1_words[i][j] in not_spam_dict:
            result2*=not_spam_dict[x1_words[i][j]]
        else:
            result2*=not_spam_dict['UNK']
    if p_spam*result1 > p_notspam*result2:
        ans.append(1)
    else:
        ans.append(0)
```

```python
len(ans)
```

### find accuracy 

```python
cnt=0
TOTAL=len(ans)
for i in range(len(ans)):
    if y1[i]==ans[i]:
        cnt+=1
accuracy=(cnt/TOTAL)*100
```
